## المسألة الأولى: توليد كلمة مرور


````python
import string

def generate_password(
    length: int,
    use_lowercase: bool = True,
    use_uppercase: bool = True,
    use_digits: bool = True,
    min_digits: int = 1,
    symbols: str = "!@#$%^&*",
    min_symbols: int = 1
) -> str:

    password = []

    # 1. نضيف الحد الأدنى من الأرقام
    if use_digits:
        for _ in range(min_digits):
            password.append(random.choice(string.digits))

    # 2. نضيف الحد الأدنى من الرموز
    for _ in range(min_symbols):
        password.append(random.choice(symbols))

    # 3. نجمع باقي الأحرف الممكنة
    all_chars = ""

    if use_lowercase:
        all_chars += string.ascii_lowercase

    if use_uppercase:
        all_chars += string.ascii_uppercase

    if use_digits:
        all_chars += string.digits

    all_chars += symbols  # نضيف الرموز أيضًا

    # 4. نكمل باقي الطول
    remaining_length = length - len(password)

    for _ in range(remaining_length):
        password.append(random.choice(all_chars))

    # 5. نخلط الباسورد عشان يكون عشوائي
    random.shuffle(password)

    # 6. نحوله لنص ونرجعه
    return ''.join(password)
````

````python
pwd = generate_password(length=10, min_digits=2, min_symbols=2)
print(pwd)
````

## المسألة الثانية: توليد عبارة مرور


````python
import random

def generate_passphrase(
    num_words: int,
    separator: str = "-",
    capitalize: bool = False,
    include_number: bool = False
) -> str:
    
    words_list = [
        "apple", "river", "moon", "train", "cloud",
        "house", "green", "light", "garden", "book",
        "sun", "star", "dream", "coffee", "flower",
        "happy", "ocean", "forest", "smile", "bird"
    ]

    # نتأكد أن عدد الكلمات صحيح
    if num_words <= 0:
        raise ValueError("Number of words must be greater than 0")

    # نختار كلمات عشوائية
    chosen_words = random.choices(words_list, k=num_words)

    # إذا نبي أول حرف كبير
    if capitalize:
        chosen_words = [word.capitalize() for word in chosen_words]

    # نجمع الكلمات بالفاصل
    passphrase = separator.join(chosen_words)

    # إذا نبي نضيف رقم
    if include_number:
        number = str(random.randint(0, 9))
        passphrase = passphrase + separator + number

    return passphrase
````

````python
result = generate_passphrase(
    num_words=4,
    separator="-",
    capitalize=True,
    include_number=True
)

print(result)
````
